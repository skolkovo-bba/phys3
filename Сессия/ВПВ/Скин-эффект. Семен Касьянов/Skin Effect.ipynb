```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем сетку для heatmap
x = np.arange(-1, 1, 0.001)
y = np.arange(-1, 1, 0.001)
X, Y = np.meshgrid(x, y)
# Вычисляем амплитуду в зависимости от расстояния до центра провода
depth = np.sqrt(X**2 + Y**2)
amplitude = [[] for _ in range(len(depth))]
for i in range (len(amplitude)):
    for j in range (len(depth[i])):
        if depth[i][j] > 1:
            amplitude[i].append(1)
        else:
            amplitude[i].append(np.exp(-((1-depth[i][j])/ 0.7)))
```

```python
# Визуализация heatmap
fig, ax = plt.subplots(figsize = (12, 8))
plt.pcolormesh(X, Y, amplitude, shading='auto', cmap='viridis')
plt.colorbar(label='Амплитуда')
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().add_patch(plt.Circle((0, 0), 1, color='black', fill=False))
plt.gca().add_patch(plt.Circle((0, 0), 0.3, color='black', fill=False))
plt.title('Распределение амплитуды волны внутри круглого провода')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
